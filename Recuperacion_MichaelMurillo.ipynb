## Tratamiento de Nulos

Algunas de estas variables contienen la definicion del null en el "documento de descripcion de los campos". Me basare en esto para rellenar algunas de las variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# % de nulos
null_percentage = df.isnull().mean() * 100
null_percentage_filtered = null_percentage[null_percentage > 0]

# Sort
null_percentage_sorted = null_percentage_filtered.sort_values(ascending=False)

# Ploteo
plt.figure(figsize=(25, 10))
sns.barplot(x=null_percentage_sorted.index, y=null_percentage_sorted, palette= 'magma')
plt.xticks(rotation=90)
plt.xlabel('Columns')
plt.ylabel('Null Percentage')
plt.title('Null Values in Each Column (Descending Order)')


for i, v in enumerate(null_percentage_sorted):
    plt.text(i, v + 0.5, f'{v:.2f}%', ha='center', va='bottom')

plt.show()
```

### Condition
Estas columnas dobles no siento que añadan mucho valor. Voy a agruparlas en una sola y reducir las categorias que contiene la variable

Artery Adjacent to arterial street <br>
Feedr Adjacent to feeder street<br>
Norm Normal<br>
RRNn Within 200' of North-South Railroad<br>
RRAn Adjacent to North-South Railroad<br>
PosN Near positive off-site feature--park, greenbelt, etc.<br>
PosA Adjacent to postive off-site feature<br>
RRNe Within 200' of East-West Railroad<br>
RRAe Adjacent to East-West Railroad<br>

```python
df.Condition1.value_counts()
```

```python
df.Condition2.value_counts()
```

Veo muchos valores con poca frecuencia. Agrupare las categorias de la siguiente manera:<br>
Norm<br>
Artery<br>
Pos = PosN & PosA<br>
RRN = RRNn & RRNe & RRAn & RRAe<br>

```python
df[['Condition1', 'Condition2']]
```

```python
def condition(row):
    if row['Condition1'] == 'Norm' and row['Condition2'] == 'Norm':
        return 'Norm'
    elif row['Condition1'] == 'Norm':
        return row['Condition2']
    elif row['Condition2'] == 'Norm':
        return row['Condition1']
    elif row['Condition1'] == row['Condition2'] and row['Condition1'] != 'Norm':
        return row['Condition1']
    else:
        return row['Condition1']

# Apply the custom function to create the 'Condition' column
df['Condition'] = df.apply(condition, axis=1)
```

```python
df.Condition.value_counts()
```

Ahora que las tengo agrupadas, voy a summarizarlas

```python
condition_map = {
    'PosN': 'Pos',
    'PosA': 'Pos',
    'RRNn': 'RRN',
    'RRNe': 'RRN',
    'RRAn': 'RRN',
    'RRAe': 'RRN',
    'Artery' : 'STR',
    'Feedr' : 'STR'
}
# Reemplazo valores
df['Condition'] = df['Condition'].replace(condition_map)
df.Condition.value_counts()
```

```python
df.drop(columns = ['Condition1', 'Condition2'], axis=0, inplace=True)
```

### 3. Total living area: Variable que indica el tamaño del espacio habitable

<b>LowQualFinSF:</b> Low quality finished square feet (all floors) <br>

Aqui encontre un problema porque queria calcular el area total construida pero hay muchas "zonas grises" por lo que me base en la siguiente pagina para determinar cuales variables sumar y cuales no:<br>

https://www.homelight.com/blog/what-is-included-in-square-footage/

En general, lo que se debe de calcular es el area total de "Living spaces". Para ello cree el siguiente calculo:<br>

BsmtFinSF1 + BsmtFinSF2 + GarageArea if GarageType = (Attchd Attached to home, Basment, BuiltIn) + 1stFlrSF + 2ndFlrSF

```python
df['Living_space'] = (df['BsmtFinSF1'] + df['BsmtFinSF2'] + df['GarageArea']) * df['GarageType'].isin(['Attchd', 'Basment', 'BuiltIn']) + df['1stFlrSF'] + df['2ndFlrSF']
```

### Lasso

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split

X = df[df.columns.difference(['SalePrice'])]
y = df.SalePrice

# Especificar una lista de valores de alpha a probar
alphas = [1, 0.1, 0.001, 0.5, 0.005, 0.5, 0.001]  # Puedes agregar más valores si lo deseas

lasso_cv = LassoCV(alphas=alphas, cv=5)  # cv=5 indica 5-fold cross-validation

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Fitting del modelo con validación cruzada para encontrar el mejor alpha
lasso_cv.fit(X_train, y_train)

# Mejor valor de alpha seleccionado por LassoCV
best_alpha = lasso_cv.alpha_

# Coeficientes del modelo con el mejor alpha
coefficients = lasso_cv.coef_

# Crear un DataFrame con las columnas y sus coeficientes
coefficients_df = pd.DataFrame({"Features_lasso": X.columns, "lasso_score": coefficients})

# Filtrar las columnas con coeficientes diferentes de cero (seleccionadas por Lasso)
lasso_scores = coefficients_df[coefficients_df["lasso_score"] != 0]

lasso_scores= lasso_scores.iloc[abs(lasso_scores["lasso_score"]).argsort()[::-1]]

print("Mejor alpha:", best_alpha)
print("Columnas seleccionadas:")
lasso_scores
lasso_scores = lasso_scores[lasso_scores['lasso_score']>0.01]
lasso_scores
```

```python
lasso_scores = lasso_scores.rename(columns={"column": "Features_lasso"})
lasso_scores = lasso_scores.drop('lasso_score', axis=1)
lasso_scores
```

```python
column_names = list(df.drop(['SalePrice'],axis=1))
# Se almacenan en df_cols
df_cols=pd.DataFrame(column_names)
df_cols['Id'] = df_cols.index
df_cols
```

```python
merged_scores = lasso_scores.merge(df_cols, left_on='Features_lasso', right_on=0)
merged_scores = merged_scores.set_index("Id")
merged_scores = merged_scores.drop(columns=[0])
lasso_scores = merged_scores
lasso_scores
```

## ElasticNet

```python
import pandas as pd
from sklearn.linear_model import ElasticNetCV
from sklearn.model_selection import train_test_split
alphas = [0.01, 0.1, 1.0, 10.0, 0.05, 0.001, 0.005] 

# Especificar la proporción de L1 (Lasso) en la penalización (l1_ratio)
l1_ratios = [0.1, 0.5, 0.7, 0.9, 0.01, 0.05] 

elastic_net_cv = ElasticNetCV(alphas=alphas, l1_ratio=l1_ratios, cv=5)  # cv=5 indica 5-fold cross-validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Fitting
elastic_net_cv.fit(X_train, y_train)

best_alpha = elastic_net_cv.alpha_
best_l1_ratio = elastic_net_cv.l1_ratio_


coefficients = elastic_net_cv.coef_
coefficients_df = pd.DataFrame({"Features_elasticnet": X.columns, "elasticnet_score": coefficients})

# Filtrar las columnas con coeficientes diferentes de cero (seleccionadas por Elastic Net)
elasticnet_scores = coefficients_df[coefficients_df["elasticnet_score"] != 0]

# Ordenar los coeficientes de mejor a peor en valor absoluto
elasticnet_scores = elasticnet_scores.iloc[abs(elasticnet_scores["elasticnet_score"]).argsort()[::-1]]

print("Mejor alpha:", best_alpha)
print("Mejor l1_ratio:", best_l1_ratio)
print("Columnas seleccionadas:")
elasticnet_scores
elasticnet_scores = elasticnet_scores[elasticnet_scores['elasticnet_score']>0.03]
elasticnet_scores
```
